```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import os
```

```python
auto = pd.read_csv("titanic.csv")
auto.head()
```

```python
sexvssurvived=auto.groupby(["sex","survived"]).size().unstack(fill_value=0)
sexvssurvived
```

```python
sexvssurvived.plot(kind='bar',stacked=True)
plt.show()
```

```python
sexVSage=auto.groupby(["sex","age"]).size().unstack(fill_value=0)
sexVSage
```

```python
sexVSage.plot(kind='bar',stacked=True)
plt.show()
```

```python
sexvsfare=auto.groupby(["sex","fare"]).size().unstack(fill_value=0)
sexvsfare
```

```python
sexvsfare.plot(kind='bar',stacked=True)
plt.show()
```
